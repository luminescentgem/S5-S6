# 1. Deviner des phrases


L'objectif de cette activité est de construire un programme qui essaie de deviner une phrase secrète.  

Tout ce que nous aurons à indiquer au programme est le nombre de lettre qu'il aura réussi à deviner mais, pas desquelles il s'agit ni où elle sont placées dans la phrase.  

Voici une phrase cible, extraite du manga _Naruto_ :

**I never go back on my word, because that is my Ninja way**

```python
target = "I never go back on my word, because that is my Ninja way."
print(len(target))
```

En anglais, il y a une multitude de caractères parmi lesquels on peut choisir : les lettres minuscules, les lettres majuscules, les espaces et la ponctuation :

```python
characters = " abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ.',?!"
```

La fonction suivante nommée `makelist()` permet de créer une liste aléatoire de caractères de la même taille que `target`. 

```python
import random

def makelist(target) :
    '''retourne une liste de charactères
    de même longuer que la cible'''
    charlist = []
    for i in range(len(target)) :
        charlist.append(random.choice(characters))
    return charlist

newlist = makelist(target)
print(newlist)
print(len(newlist))
```

Nous venons de céer une liste de caractères de même longueur que `target`.  

Pourquoi créer une liste plutôt qu'une de chaîne de caractère ?  

Les listes sont souvent plus faciles à manipuler que les chaînes de caractères. Par exemple, il n'est pas possible de remplacer (simplement) un caractère par un autre dans une chaîne de caractère.

Pour revenir à une chaine de caractères, on pourra utiliser la méthode `join()` :

```python
"".join(newlist)
```
